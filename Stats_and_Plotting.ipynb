Importing the stats subcategory from the scipy library

```python
from scipy import stats
```

Tests if the population mean would be equal to a specific value.  With a p-value of 10^-28 we can claim that the population mean for the IQ (VIQ measure) is not 0.

```python
stats.ttest_1samp(data['VIQ'], 0)   
```

Averaging the mean VIQs of the two gender groups and testing if they are the same. We can claim the group VIQs are the same

```python
female_viq = data[data['Gender'] == 'Female']['VIQ']
male_viq = data[data['Gender'] == 'Male']['VIQ']
stats.ttest_ind(female_viq, male_viq)   
```

Let us test if FISQ and PIQ are significantly different. We can use a 2 sample test:

```python
stats.ttest_ind(data['FSIQ'], data['PIQ'])   
```

FSIQ and PIQ are measured on the same individuals. Thus the variance due to inter-subject variability is confounding, and can be removed, using a “paired test”, or “repeated measures test”:

```python
stats.ttest_rel(data['FSIQ'], data['PIQ'])   
```

This is equivalent to a 1-sample test on the difference:

```python
stats.ttest_1samp(data['FSIQ'] - data['PIQ'], 0)   
```

T-tests assume Gaussian errors. We can use a Wilcoxon signed-rank test, that relaxes this assumption:

```python
stats.wilcoxon(data['FSIQ'], data['PIQ'])   
```

Testing if the weights between the gender groups is the same. According to the small p value from the t test, their average weights are different.

```python
female_viq = data[data['Gender'] == 'Female']['Weight']
male_viq = data[data['Gender'] == 'Male']['Weight']
stats.ttest_ind(female_viq, male_viq, nan_policy='omit')   
```

Testing whether the VIQs between the gender groups is the same with a non parametric statistical test (Mann-Whitney U Test). the high p value indicates that they are the same

```python
female_viq = data[data['Gender'] == 'Female']['VIQ']
male_viq = data[data['Gender'] == 'Male']['VIQ']
stats.mannwhitneyu(female_viq, male_viq, nan_policy='omit')   
```

```python

if not os.path.exists('wages.txt'):
    # Download the file if it is not present
    urllib.request.urlretrieve('http://lib.stat.cmu.edu/datasets/CPS_85_Wages',
                       'wages.txt')

# Give names to the columns
names = [
    'EDUCATION: Number of years of education',
    'SOUTH: 1=Person lives in South, 0=Person lives elsewhere',
    'SEX: 1=Female, 0=Male',
    'EXPERIENCE: Number of years of work experience',
    'UNION: 1=Union member, 0=Not union member',
    'WAGE: Wage (dollars per hour)',
    'AGE: years',
    'RACE: 1=Other, 2=Hispanic, 3=White',
    'OCCUPATION: 1=Management, 2=Sales, 3=Clerical, 4=Service, 5=Professional, 6=Other',
    'SECTOR: 0=Other, 1=Manufacturing, 2=Construction',
    'MARR: 0=Unmarried,  1=Married',
]

short_names = [n.split(':')[0] for n in names]

data = pandas.read_csv('wages.txt', skiprows=27, skipfooter=6, sep=None,
                       header=None)
data.columns = short_names

# Log-transform the wages, because they typically are increased with
# multiplicative factors
import numpy as np
data['WAGE'] = np.log10(data['WAGE'])
```
